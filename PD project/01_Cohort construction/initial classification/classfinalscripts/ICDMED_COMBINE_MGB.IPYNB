```python
import pandas as pd

def merge_patient_ids_efficient(file_paths):
    unique_patient_ids = set()
    
    for file_path in file_paths:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract patient IDs and add them to the set
        patient_ids = df['BDSPPatientID']
        unique_patient_ids.update(patient_ids)
    
    return unique_patient_ids

## ICD


file_paths = ['ICDMIN_MGB_1_NEW.csv', 'ICDMIN_MGB_2_NEW.csv', 'ICDMIN_MGB_3_NEW.csv', 'ICDMIN_MGB_4_NEW.csv']
unique_ids_ICD_minus = merge_patient_ids_efficient(file_paths)
print(f"Total icdminus unique patients: {len(unique_ids_ICD_minus)}")
# print(unique_patient_ids)

# type(unique_patient_ids_ICDMIN)

# print(unique_patient_ids_ICDMIN)

file_paths = ['ICDPL_MGB_New.csv']
unique_ids_ICD_plus = merge_patient_ids_efficient(file_paths)
print(f"Total icdplus unique patients: {len(unique_ids_ICD_plus)}")
# print(unique_patient_ids)


common_patient_ids_before = unique_ids_ICD_plus.intersection(unique_ids_ICD_minus)
print('\n amount common patients before')
print(len(common_patient_ids_before))

# exlude overlapping Ids from mimus
unique_ids_ICD_minus = unique_ids_ICD_minus.difference(unique_ids_ICD_plus)

# evaluate
common_patient_ids_after= unique_ids_ICD_plus.intersection(unique_ids_ICD_minus)
print('\n amount common patients after')
print(len(common_patient_ids_after))

print(f"Total icdminus unique patients: {len(unique_ids_ICD_minus)}")
print(f"Total icdplus unique patients: {len(unique_ids_ICD_plus)}")


```
